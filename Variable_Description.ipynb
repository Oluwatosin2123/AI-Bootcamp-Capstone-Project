```python
# Import the StandardScaler class from scikit-learn
from sklearn.preprocessing import StandardScaler
```

```python
# First, check if there are any numeric columns
numeric_cols = VariableDescription_data.select_dtypes(include=['int64', 'float64']).columns.tolist()
print(f"Numeric columns found: {numeric_cols}")  # Debug print to see what columns were found

# If no numeric columns are found automatically, you might need to convert some columns
# For example, if you know certain columns should be numeric:
# columns_to_convert = ['column1', 'column2']  # Replace with actual column names
# for col in columns_to_convert:
#     VariableDescription_data[col] = pd.to_numeric(VariableDescription_data[col], errors='coerce')

# Now, identify and handle non-numeric values in the numeric columns
if numeric_cols:  # Only proceed if there are numeric columns
    for col in numeric_cols:
        # Replace problematic strings with NaN
        VariableDescription_data[col] = pd.to_numeric(VariableDescription_data[col], errors='coerce')
        
        # Fill NaN values with an appropriate value (median is often a good choice)
        # Check if there are non-NaN values before calculating median
        if VariableDescription_data[col].notna().any():
            VariableDescription_data[col] = VariableDescription_data[col].fillna(VariableDescription_data[col].median())
        else:
            VariableDescription_data[col] = VariableDescription_data[col].fillna(0)  # Default to 0 if all NaN

    # Now apply the StandardScaler only if there are numeric columns
    scaler = StandardScaler()
    VariableDescription_data[numeric_cols] = scaler.fit_transform(VariableDescription_data[numeric_cols])
else:
    print("No numeric columns found for scaling. Please check your data.")
```

```python
#....Categorical Colunms

categorical_cols =  VariableDescription_data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\n VALUE COUNTS FOR: {col}")
    print( VariableDescription_data[col].value_counts().head(10))
```
